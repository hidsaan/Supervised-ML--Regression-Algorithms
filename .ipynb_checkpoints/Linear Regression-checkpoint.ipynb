# Dataset 1: 1000 Companies

```python
import pandas as pd
```

```python
#Step 1: Fetch the data
data= pd.read_csv('1000_Companies.csv')
data.head()
```

```python
#Step 2: Analyse the data and perform data cleaning
data.info()
```

```python
#Step 3: Choose model- a) Choose X- independent variable/ variables (input):
X=data.iloc[:,[0,1,2,3]].values
X
```

```python
#b) Choose dependent variable (output):
Y= data.iloc[:,4].values
Y
```

```python
#Step 4: In order to change the object values(string values) into computer undertandable int format, we need to use LabelEncoder from Skikit:
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
X[:,3]=lb.fit_transform(X[:,3]) #Turns the names of the cities into numbers
X
```

```python
#Step 5: Train and test set up using train_test_split from Skikit:
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2) #test_size didtributes the values to be tested and trained, eg here, 80% are traines and 20% are tested
```

```python
#Step 6: Training the algo using linear regression- Giving the proper regression algorithm based on data:
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
#Step 7: Checking the accuracy:
accuracy=regressor.score(X_test,Y_test)
accuracy
```

```python
#Step 8: Testing the algo- letting it predict the dependent variable by entering the independent variables:
new_entry= [170000.8, 190220.4, 417000.1, 'New York']
new_entry[3]= lb.fit_transform([new_entry[3]])[0] #transform into int
```

```python
predict= regressor.predict([new_entry])
print('The Profit on Investment is: ', predict)
```

# Dataset 2: Fish

```python
my_data=pd.read_csv('Fish.csv')
my_data.head()
```

```python
my_data.info()
```

```python
my_data.shape
```

```python
x=my_data.iloc[:,[2,3,4,5,6]].values
y=my_data.iloc[:,1].values
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y, test_size=0.3)
y_test
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
regressor.score(x_test,y_test)
```

```python
predict=regressor.predict([[22.1,23.5,26.8,7.3968,4.1272]])
print('The weight of the fish is: ',predict)
```
