# Dataset 1: Mileage

```python
import pandas as pd
```

```python
data=pd.read_csv('auto-mpg.csv')
data.head()
```

```python
data.info()
```

```python
X=data.iloc[:,[1,2,3,5,6]].values
Y=data.iloc[:,0].values
X
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
X[:,2]=lb.fit_transform(X[:,2])
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(X,Y, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_train
```

```python
from sklearn import svm
classifier=svm.SVR(kernel='linear')
classifier.fit(X_train,Y_train)
```

```python
classifier.score(X_test,Y_test)
```

```python
new_data=[8,307.0,'130',12.0,70]
new_data[2]=lb.transform([new_data[2]])[0]
new_data
```

```python
predict=classifier.predict([new_data])
print('The milege of the car given in Miles Per Gallon is: ', predict)
```

# Dataset 2: Wine Quality

```python
my_data=pd.read_csv('WineQT.csv')
my_data.head()
```

```python
my_data.info()
```

```python
x=my_data.iloc[:,[0,1,3,4,6,7,8,9,10]].values
y=my_data.iloc[:,11].values
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
x_train
```

```python
from sklearn import svm
regressor=svm.SVR(kernel='rbf')
regressor.fit(x_train,y_train)
```

```python
regressor.score(x_test,y_test)
```

```python
result=regressor.predict([[7.4,0.70,1.9,0.076,34.0,0.9978,3.51,0.56,9.4]])
print('The rating out of 10 is: ',result)
```
