# Dataset 1: Car Price

```python
import pandas as pd
```

```python
#Fetch and analyze:
my_data=pd.read_csv('cardekho.csv')
my_data.head()
```

```python
my_data.info()
```

```python
#Data Cleaning:
clean_data=my_data.dropna()
clean_data.info()
```

```python
#Choosing Model:
X=clean_data.iloc[:,[1,3,4,6]].values
Y=clean_data.iloc[:,2].values
X
```

```python
#Transformation of more than one columns:
from sklearn.preprocessing import LabelEncoder

#Create label for first column:
lb=LabelEncoder()
X[:,2]=lb.fit_transform(X[:,2])

#Create label for second column:
lb1= LabelEncoder()
X[:,3]=lb1.fit_transform(X[:,3])
X
```

```python
#Proceed with testing and training as usual:
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test= train_test_split(X,Y, test_size=0.2)
Y_test
```

```python
#Solving the Algorith using Forest Regression as it has 2 categorical values as independent variables:
from sklearn.ensemble import RandomForestRegressor

regressor=RandomForestRegressor(n_estimators=300)
regressor.fit(X_train, Y_train)
```

```python
#Checking the accuracy:
accuracy= regressor.score(X_test, Y_test)
accuracy
```

```python
#Entering new data:
add_new=[2019, 15500, 'LPG', 'Manual']

#Transforming the third and 4th column one by one according to the label assigned to them above already:
add_new[2] = lb.transform([add_new[2]])[0]  # Transform using the fitted encoder for the third column
add_new[3] = lb1.transform([add_new[3]])[0]

add_new
```

```python
#Prediction Test:
new_predict=regressor.predict([add_new])
print('The Predicted Car Selling price:',new_predict)
```

# Dataset 2: Concrete Compressive Strength

```python
my_data=pd.read_csv('Concrete_Data.csv')
my_data.head()
```

```python
my_data.info()
```

```python
x=my_data.iloc[:,[0,1,2,3,4,5,6,7]].values
y=my_data.iloc[:,8].values
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y, test_size= 0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor()
regressor.fit(x_train,y_train)
```

```python
regressor.score(x_test,y_test)
```

```python
result=regressor.predict([[540.0,0.0,0.0,162.0,2.5,1040.0,676.0,28]])
print('The compressive strength of concrete is:', result)
```
