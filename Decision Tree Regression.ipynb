# Dataset 1: Position Salaries

```python
import pandas as pd
```

```python
data=pd.read_csv('Position_Salaries.csv')
data.head()
```

```python
X=data.iloc[:,1:2].values
Y=data.iloc[:,2].values
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(X,Y, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr= DecisionTreeRegressor()
dtr.fit(X_train,Y_train)
```

```python
dtr.score(X_test,Y_test)
```

```python
dtr.predict([[3.4]])
```

# Dataset 2: Housing Prices

```python
my_data=pd.read_csv('housing.csv')
my_data.head()
```

```python
my_data.info()
```

```python
clean_data=my_data.dropna()
clean_data.info()
```

```python
x=clean_data.iloc[:,[0,1,2,3,6,7,9]].values
y=clean_data.iloc[:,8].values
x
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
x[:,6]=lb.fit_transform(x[:,6])
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor()
regressor.fit(x_train,y_train)
```

```python
regressor.score(x_test,y_test)
```

```python
new_data=[-112.23,39.88,42.0,780.0,116.0,8.3252,'INLAND']
new_data[6]=lb.transform([new_data[6]])[0]

result=regressor.predict([new_data])
print('the mean value of the house is:', result)
```
